```python
import numpy as np
import pandas as pd
from __future__ import print_function
from __future__ import division
```

```python
baseball = pd.read_csv("../data/baseball.csv", index_col='id')
baseball.head()
```

For the equation of OBP, 

$$OBP = \frac{H + BB + HBP}{AB + BB + HBP + SF}$$

If AB + BB + HBP + SF = zero, it will cause error because it cannot divide by zero, so they will be NaN.

```python
def obp(x):
    if x['ab'] + x['bb'] + x['hbp'] + x['sf'] == 0:
        return np.NaN
    else:
        return (x['h'] + x['bb'] + x['hbp'])/(x['ab'] + x['bb'] + x['hbp'] + x['sf'])
```

```python
baseball['OBP'] = baseball.apply(obp, axis = 1)
baseball[['player','year','OBP']]
```
